```python
import pandas as pd
import numpy as np
import textwrap
from transformers import pipeline
```

```python
# Read in our data
df = pd.read_csv('bbc_text_cls.csv')
```

```python
# Inspect the data
df.head()
```

```python
# Pick an random article from business class
doc = df[df.labels == 'business'] ['text'].sample(random_state=42)
```

```python
# define a textwrapping function, the text is so long that it will go of the page
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
# Print the selected article
print(wrap(doc.iloc[0]))
```

```python
# Instansiate a summarizer pipline object, using a distilbart model
summarizer = pipeline("summarization")
```

```python
# pass the article into the summarizer created above
summarizer(doc.iloc[0].split("\n", 1)[1])
```

```python
# function to do the above and print out the result
def print_summary(doc):
  result = summarizer(doc.iloc[0].split("\n", 1)[1])
  print(wrap(result[0]['summary_text']))
```

```python
# Seems to be a good summary that tells the essence of the article
print_summary(doc)
```

```python
# Select another random article
doc = df[df.labels == 'entertainment']['text'].sample(random_state=123)
print(wrap(doc.iloc[0]))
```

```python
# summarize the document
print_summary(doc)
```
